```python
# libraries and data
import matplotlib.pyplot as plt
import numpy as np
 
# Make a data frame
#df=pd.DataFrame({'x': y_data, 'SQFT_LIVING': x_data['sqft_living'],'SQFT_LOT': x_data['sqft_lot'],'SQFT_ABOVE': x_data['sqft_above']})
df=pd.DataFrame({'x': y_data, 'SQFT_LIVING': x_data['sqft_living'],'SQFT_ABOVE': x_data['sqft_above']})

 
# style
plt.style.use('seaborn-darkgrid')
 
# create a color palette
palette = plt.get_cmap('Set1')
 
# multiple line plot
num=0
for column in df.drop('x', axis=1):
    num+=1
    plt.plot(df['x'], df[column], marker='', color=palette(num), linewidth=1, alpha=0.9, label=column)
 
# Add legend
plt.legend(loc=2, ncol=2)
 
# Add titles
plt.title("A Spaghetti plot", loc='left', fontsize=12, fontweight=0, color='orange')
plt.xlabel("Price")
plt.ylabel("SQFT")
```

```python
#Formula for normal equation
theta_best = np.linalg.inv(x_data.T.dot(x_data)).dot(x_data.T).dot(y_data)
```
